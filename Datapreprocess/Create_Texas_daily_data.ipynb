```python
import pandas as pd
from functools import reduce
```

```python
#combine all the Texas hourly usage data in 2015
texas_data1_6 = pd.read_csv('texas-hourly-1-6.csv', na_values=[])
texas_data6_9 = pd.read_csv('texas-hourly-6-9.csv', na_values=[])
texas_data9_10 = pd.read_csv('901-1001_2015all.csv', na_values=[])
texas_data10_11 = pd.read_csv('1001-1101_2015all.csv', na_values=[])
texas_data11_12 = pd.read_csv('1101-1201_2015all.csv', na_values=[])
texas_data12_1 = pd.read_csv('1201-0101_2015all.csv', na_values=[])
```

```python
#collect Texas data ID without missing value
dataid = reduce(set.intersection,map(set,[texas_data1_6['localhour'],texas_data6_9['localhour'],texas_data9_10['dataid'],texas_data10_11['dataid'],texas_data11_12['dataid'],texas_data12_1['dataid']]))
```

```python
texas_data1_6 = texas_data1_6.rename(columns=lambda x: x.replace('\t', ''))
texas_data1_6 = texas_data1_6.rename(columns={'dataid':'localhour','localhour':'dataid'})
texas_data1_6.replace('\n','')
texas_data1_6 = texas_data1_6[texas_data1_6['dataid'].isin(dataid)]
texas_data1_6.tail()
```

```python

texas_data6_9 = texas_data6_9.rename(columns=lambda x: x.replace('\t', ''))
texas_data6_9 = texas_data6_9.rename(columns={'dataid':'localhour','localhour':'dataid'})
texas_data6_9.replace('\n','')
texas_data6_9 = texas_data6_9[texas_data6_9['dataid'].isin(dataid)]
texas_data6_9.tail()
```

```python

texas_data9_10 = texas_data9_10.rename(columns=lambda x: x.replace('\t', ''))
texas_data9_10.replace('\n','')
texas_data9_10 = texas_data9_10[texas_data9_10['dataid'].isin(dataid)]
texas_data9_10.tail()
```

```python

texas_data10_11 = texas_data10_11.rename(columns=lambda x: x.replace('\t', ''))
texas_data10_11.replace('\n','')
texas_data10_11 = texas_data10_11[texas_data10_11['dataid'].isin(dataid)]
texas_data10_11.tail()
```

```python

texas_data11_12 = texas_data11_12.rename(columns=lambda x: x.replace('\t', ''))
texas_data11_12.replace('\n','')
texas_data11_12 = texas_data11_12[texas_data11_12['dataid'].isin(dataid)]
texas_data11_12.tail()
```

```python

texas_data12_1 = texas_data12_1.rename(columns=lambda x: x.replace('\t', ''))
texas_data12_1.replace('\n','')
texas_data12_1 = texas_data12_1[texas_data12_1['dataid'].isin(dataid)]
texas_data12_1.tail()
```

```python
Texas_data = [texas_data1_6, texas_data6_9, texas_data9_10, texas_data10_11, texas_data11_12, texas_data12_1]
Texas_2015 = pd.concat(Texas_data)
```

```python
Texas_2015.fillna(0,inplace= True)
```

```python
import gc
del texas_data1_6
del texas_data6_9
del texas_data9_10
del texas_data10_11
del texas_data11_12
del texas_data12_1
gc.collect()
```

```python
import pandas as pd
import numpy as np
hourly_data_PSTexas=pd.read_pickle('hourly_data_PSTexas.pickle')
```

```python
#remove data with missing value
train={}
for i in dataid:
    if hourly_data_PSTexas[hourly_data_PSTexas['dataid']==i].sort_values(by='Localhour').values[:,2:].shape == (8814, 15):
        train[i]=(hourly_data_PSTexas[hourly_data_PSTexas['dataid']==i].sort_values(by='Localhour').values[:,2:])
   
```

```python
np.save('./PreprocessedData/hourly_data_PSTexas_trainingset',np.array(list(train.values())))
```

```python
np.save('./PreprocessedData/hourly_data_PSTexas_HouseholdID',np.array(list(train.keys())))
```

```python
hourly_data_PSTexas_trainingset=np.load('./PreprocessedData/hourly_data_PSTexas_trainingset.npy')
```

```python
#calculate the daily data in Texas and save
daily_data_PSTexas_data=[]
for j in range(len(train)):
    first = np.sum(hourly_data_PSTexas_trainingset[j,0:24,:],axis=0)
    for i in range(1,365):
        first = np.vstack([first,np.sum(hourly_data_PSTexas_trainingset[j,i*24:(i+1)*24,:],axis=0)])
    daily_data_PSTexas_data.append(first)
daily_data_PSTexas_data=np.array(daily_data_PSTexas_data)
np.save('./PreprocessedData/daily_data_PSTexas_data.npy',daily_data_PSTexas_data)
```
